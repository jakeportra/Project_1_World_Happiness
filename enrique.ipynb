```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# read in the csv
data = pd.read_csv("Resources/Average annual hours actually worked.csv")

# drop all empty columns
data = data.dropna("columns", how="all")

# only total employment
total_employment = data.loc[data["Employment status"] == "Total employment"]
```

```python
# pull out columns we want and rename them
data = total_employment[["Country", "Time", "Value"]]
data = data.rename(columns={"Time": "Year", "Value": "Avg Hours"})

# just 2018 data
data_2018 = data.loc[data["Year"] == 2018]

# sort in alphabetical order by country
data_2018 = data_2018.sort_values(by="Country")
```

```python
# plots

x_axis = data_2018["Country"].tolist()
y_axis = data_2018["Avg Hours"].tolist()

plt.plot(x_axis, y_axis)

plt.show()
```
