```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
# read in the csv
data = pd.read_csv("Resources/Average annual hours actually worked.csv")

# drop all empty columns
data = data.dropna("columns", how="all")

# total employment and dependent employment
dependent_employment = data.loc[data["Employment status"] == "Dependent employment"]
total_employment = data.loc[data["Employment status"] == "Total employment"]
```

```python
# just 2018 data
data_2018 = data.loc[data["TIME"] == 2018]

# total and dependent data 2018
dependent_2018 = data_2018.loc[data["Employment status"] == "Dependent employment"]
total_2018 = data_2018.loc[data["Employment status"] == "Total employment"]

# sort in alphabetical order by country
data_2018 = data_2018.sort_values(by="Country")

print(len(total_2018))
print(len(dependent_2018))

data_2018
```

```python
# plots

# create lists with axis data
total_avg = data_2018.loc[data_2018["Employment status"] == "Total employment", "Value"]

# create labels
labels = data_2018["Country"].unique().tolist()

# create tick locations and rotate 45 degrees
x = np.arange(len(labels))
plt.xticks(x, labels, rotation=45)

# plot the data
plt.bar(labels, total_avg)

plt.tight_layout()
plt.show()
```

```python
# plots

# create labels
labels = data_2018["Country"].tolist()

# create lists with axis data
dependent_avg = data_2018.loc[data_2018["Employment status"] == "Dependent employment", "Value"]
total_avg = data_2018.loc[data_2018["Employment status"] == "Total employment", "Value"]

# create tick locations and rotate 45 degrees
x = np.arange(len(labels))
width = 0.35

# plot the data
fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, dependent_avg, width, label="Dependent Employment")
rects2 = ax.bar(x + width/2, total_avg, width, label="Total Employment")

ax.set_title("")
ax.set_ylabel("")
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

"""
def autolabel(rects):
    for rect in rects:
        height - rect.get_height()
        ax.annotate('{}'.format(height),
                   xy=(rect.get_x() + rect.get_width() / 2, height),
                   xytext=(0,3),
                   textcoords="offset points",
                   ha='center', va='bottom')
"""

autolabel(rects1)
autolabel(rects2)
fig.tight_layout()
plt.show()
```
